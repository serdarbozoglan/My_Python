```python
import pandas as pd
```

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.tail()
# NaN means "not a number" means missing values
```

```python
ufo.isnull().tail()
## Null olanlari True olarak gosterir
```

```python
ufo.notnull().tail()
## Null olmayanlari True olarak gosterir
```

```python
ufo.isnull().sum()
## Number of missing values in each column
## Burada True (missing value olanlar (True) 1 olarak, dolu hucreleri (False) 0 olarak int degerine cevirerek toplar)
```

```python
ufo.City.isnull().sum()
```

```python
ufo[ufo.City.isnull()]
```

```python
## METHOD 1 : DROPPING MISSING VALUES
ufo.shape
```

```python
ufo.dropna(how="any").shape
## DROPS ROWS WHICH HAS ANY MISISNG VALUE. Bir satirda herhangi bir bos veri varsa satiri dusurur
## dropna() gecici olarak dusurur satirlari orihjinal DF etkilenmez atama yapilmazsa
```

```python
ufo.dropna(how="all").shape 
## DROPS ROWS IF ALL VALUES ARE MISSING. Bir satirdaki tum veriler bossa satiri siler
```

```python
ufo.dropna(subset = ["City", "Shape Reported"], how="any").shape
## Either City VEYA Shape Reported dan herhangi birisnde missing value varsa dusurur
```

```python
ufo.dropna(subset = ["City", "Shape Reported"], how="all").shape
## BOTH City VE Shape Reported dan her ikisinde de missing value varsa dusurur
```

```python
ufo["Shape Reported"].value_counts()
## how many times this values counted in this column
## DEFAULT : MISSING VALUES ARE EXCLUDED
```

```python
ufo["Shape Reported"].value_counts(dropna=False)
## THIS INCLUDES NaN (MISSING VALUES) AS WELL
```

```python
ufo["Shape Reported"].fillna(value="VARIOUS", inplace = True)
## missing value lari VARIOUS degeri ile degistirir ve tablonun uzerine kaydeder/degisitirir DF yi cunku 
## inplace = True oldugu icin. DEFAULT degeri inplace= False tur o yuzden orijinal DF yi degistiremez
ufo["Shape Reported"].value_counts(dropna=False)
## bos hucreler yerine VARIOUS isimli deger gelmistir NaN yerine
```

```python
# HOW TO CREATE DUMMY(INDICATOR) VARIABLES IN PANDAS. CATEGORICAL FEATURE ICIN KULLANILIR

train = pd.read_csv("http://bit.ly/kaggletrain")
train.head()
```

```python
## ILK YONTEM : SERIES MAP METHOD 

train["Sex_male"] = train.Sex.map({"female":0, "male":1})
## "Sex_male" isimlli bir feature olustururuz. train.Sex serisi ne map methodunu dictionary uzerinden 
## kullanarak female lari 0, male_lari 1 olarak ifade ederiz ve Sex_male sutunu ile sadece erkeleri 1 ile 
## markalayarak male ve femalelari anlayabiliriz
```

```python
train.head()
## sona ekledi yeni sutunu
```

```python
## IKINCI METHOD get_dummies()


pd.get_dummies(train.Sex)
## tum olasi kategorik degerler icin bir sutun olusturur. Her satir icin hangi kategorik degere ait oldugunu belirtir. 
## toplamda k kadar olasi kategorik deger varsa k-1 tane dummy variable sutunu olustururuz. Bu ornek icin female sutunu
## dusurmemiz yeteri 
```

```python
pd.get_dummies(train.Sex).iloc[:,1:]
## bu kod ile tum satirlari ve sadece tek sutunu (2-1=1 sutun k=2 icin (female ve male k degerleri)) dahil etmemiz yeterlidir
## burada male ise 1 degilse 0 degerini alir.
```

```python
pd.get_dummies(train.Sex, prefix = "Sex").iloc[:,1:0]
## prefix ile olsuturulan sutunun basliginin Sex_ olmasi saglanir. _ degeri default olark gelir.
## 3 Kategorili bir ornek icin DataSchool How do I create dummy variable in pandas? isimli Kevin Markahm Youtube videosu...
```

```python
## Iki DataFrame i birlestirme
## embarked_dummies e 3 kategorili bir DataFrameden 2 kategorisini atariz

embarked_dummies = pd.get_dummies(train.Embarked).iloc[:,1:]
```

```python
## pd.concat ile DF leri birlestiririz ve tekrar train e atayarak overwrite ederiz

train = pd.concat([train, embarked_dummies], axis=1)
```

```python
train2 = pd.read_csv("http://bit.ly/kaggletrain")
pd.get_dummies(train2, columns = ["Sex", "Embarked"])
##train DF de:
   ##- Sex ve Embarked sutunlarini dummy sutunlara ayirir ve orijinal DF deki Sex ve Embarked suutnlarini dusurur""" 
```
