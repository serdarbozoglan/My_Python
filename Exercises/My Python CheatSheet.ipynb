# Generator

```python
# Generators more memory efficient instead of creating a list

def get_cubes(n):
    
    for i in range(n):
        yield i**3 # yield buradaki anahtar kelime, dikkat et rerun yok
```

```python
get_cubes(10)
```

```python
for x in get_cubes(10):
    print ((x))
```

```python
list(get_cubes(10))
```

```python
# Fibonacci Series with generators
```

```python
def get_fibonacci(n):
    a=1
    b=1
    
    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
for i in get_fibonacci(10):
    print(i)
```

## PRETTY PRINT (PPRINT)

```python
import pprint
stuff = ['spam', 'eggs', 'lumberjack', 'knights', 'ni']
stuff.insert(0, stuff[:])
```

```python
print(stuff)
```

```python
## indent 4 char
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(stuff)
```

```python
## 41 char per line without breaking any words
pp = pprint.PrettyPrinter(width=41, compact=True)
pp.pprint(stuff)
```

```python
tup = ('spam', ('eggs', ('lumberjack', ('knights', ('ni', ('dead',
... ('parrot', ('fresh fruit',))))))))
pp = pprint.PrettyPrinter(depth=6) # depth level is 6
pp.pprint(tup)
```
