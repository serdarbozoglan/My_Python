### second question

```python
symbols = ['AAPL', 'TSLA']
```

```python
def get_data(symbol):
    if symbol == 'AAPL':
        return [100, 101, 102]
    else:
        return [200, 201, 202]
```

```python
get_data('AAPL')
```

### Dictionary Comprehension


```python
# {k:v for x in something} key migh be x or whatever the relation may be value is x
```

```python
d = {symbol:get_data(symbol) for symbol in symbols}
d 
```

```python
pd.DataFrame(d)
```

### Second way with pd.Series

```python
l = [pd.Series(get_data(symbol), name=symbol) for symbol in symbols]
```

```python
l[0]
# Name is AAPL
```

```python
pd.concat(l, axis='columns')
```
